```python
import pandas as pd


file_path = 'telecom_churn.csv'
telecom_data = pd.read_csv(file_path)

print(telecom_data.head())
```

Overall Churn Rate


```python

churn_rate = telecom_data['Churn'].value_counts(normalize=True) * 100
print("Overall Churn Rate:")
print(churn_rate)
```

Churn Rate by Contract Renewal

```python
#by ContractRenewal
churn_rate_by_contract = telecom_data.groupby('ContractRenewal')['Churn'].value_counts(normalize=True).unstack() * 100
print("Churn Rate by Contract Renewal:")
print(churn_rate_by_contract)

# by DataPlan
churn_rate_by_dataplan = telecom_data.groupby('DataPlan')['Churn'].value_counts(normalize=True).unstack() * 100
print("Churn Rate by Data Plan:")
print(churn_rate_by_dataplan)

# by Customer Service Calls
churn_rate_by_custservcalls = telecom_data.groupby('CustServCalls')['Churn'].value_counts(normalize=True).unstack() * 100
print("Churn Rate by Customer Service Calls:")
print(churn_rate_by_custservcalls)
```

Correlation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt


corr_matrix = telecom_data.corr()


plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```

Detailed Analysis by Key Features:
Explore the relationship between churn and key features in detail.

Customer Service Calls



```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Churn', y='CustServCalls', data=telecom_data)
plt.title('Customer Service Calls by Churn')
plt.xlabel('Churn')
plt.ylabel('Customer Service Calls')
plt.show()
```

Monthly Charges

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='Churn', y='MonthlyCharge', data=telecom_data)
plt.title('Monthly Charge by Churn')
plt.xlabel('Churn')
plt.ylabel('Monthly Charge')
plt.show()
```

Day Minutes

```python

plt.figure(figsize=(12, 6))
sns.boxplot(x='Churn', y='DayMins', data=telecom_data)
plt.title('Day Minutes by Churn')
plt.xlabel('Churn')
plt.ylabel('Day Minutes')
plt.show()
```
