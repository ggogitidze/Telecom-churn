```python
import pandas as pd

file_path = 'telecom_churn.csv'
telecom_data = pd.read_csv(file_path)

```

Distribution of Monthly Charge

```python
import seaborn as sns
import matplotlib.pyplot as plt

# blue for not churned, orange for churned
colors = ['#1f77b4', '#ff7f0e']  

plt.figure(figsize=(12, 6))
sns.histplot(data=telecom_data, x='MonthlyCharge', hue='Churn', multiple='stack', kde=True, palette=colors)
plt.title('Distribution of Monthly Charge by Churn')
plt.xlabel('Monthly Charge')
plt.ylabel('Count')
plt.legend(title='Churn', labels=['Not Churned', 'Churned'])
plt.show()
```
