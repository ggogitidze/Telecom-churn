```python
import pandas as pd

file_path = 'telecom_churn.csv'
telecom_data = pd.read_csv(file_path)

```

Churn Trend Over Time

```python
import matplotlib.pyplot as plt


churn_trend = telecom_data.groupby('AccountWeeks')['Churn'].value_counts(normalize=True).unstack().fillna(0)
churn_trend['Churn Rate'] = churn_trend[1] / (churn_trend[1] + churn_trend[0])

plt.figure(figsize=(12, 6))
plt.plot(churn_trend.index, churn_trend['Churn Rate'], marker='o')
plt.title('Churn Rate Trend Over Account Weeks')
plt.xlabel('Account Weeks')
plt.ylabel('Churn Rate')
plt.grid(True)
plt.show()
```

Service Usage Trends for Churned vs. Retained Customers:

```python
# churned vs. retained customers
data_usage_trend = telecom_data.groupby(['AccountWeeks', 'Churn'])['DataUsage'].mean().unstack()

plt.figure(figsize=(12, 6))
plt.plot(data_usage_trend.index, data_usage_trend[0], label='Retained', marker='o')
plt.plot(data_usage_trend.index, data_usage_trend[1], label='Churned', marker='o')
plt.title('Average Data Usage Trend Over Account Weeks')
plt.xlabel('Account Weeks')
plt.ylabel('Average Data Usage')
plt.legend()
plt.grid(True)
plt.show()
```

Customer Service Calls Trend:


```python
#churned vs. retained customers
cust_serv_calls_trend = telecom_data.groupby(['AccountWeeks', 'Churn'])['CustServCalls'].mean().unstack()


plt.figure(figsize=(12, 6))
plt.plot(cust_serv_calls_trend.index, cust_serv_calls_trend[0], label='Retained', marker='o')
plt.plot(cust_serv_calls_trend.index, cust_serv_calls_trend[1], label='Churned', marker='o')
plt.title('Average Customer Service Calls Trend Over Account Weeks')
plt.xlabel('Account Weeks')
plt.ylabel('Average Customer Service Calls')
plt.legend()
plt.grid(True)
plt.show()
```

Monthly Charges Trend:

```python
# churned vs. retained customers
monthly_charge_trend = telecom_data.groupby(['AccountWeeks', 'Churn'])['MonthlyCharge'].mean().unstack()

plt.figure(figsize=(12, 6))
plt.plot(monthly_charge_trend.index, monthly_charge_trend[0], label='Retained', marker='o')
plt.plot(monthly_charge_trend.index, monthly_charge_trend[1], label='Churned', marker='o')
plt.title('Average Monthly Charge Trend Over Account Weeks')
plt.xlabel('Account Weeks')
plt.ylabel('Average Monthly Charge')
plt.legend()
plt.grid(True)
plt.show()
```
